```python
import h5py

# Path to the HDF5 file
file_path = '../datasets/train_0.h5'

# Function to print the structure of the HDF5 file
def print_structure(name, node):
    print(f"{name}: {node}")

# Open the HDF5 file and list its contents
with h5py.File(file_path, 'r') as hdf5_file:
    count = [0]  # Use a list to maintain state across recursive calls
    
    def visit_func(name, node):
        if count[0] < 100:
            print_structure(name, node)
            count[0] += 1

    hdf5_file.visititems(visit_func)
```

```python
def read_data(file_path, num_configs=25):
    properties = {
        "atomic_numbers": [],
        "cell": [],
        "charges": [],
        "energy": [],
        "forces": [],
        "positions": [],
        "stress": [],
        "virials": [],
    }

    with h5py.File(file_path, 'r') as hdf5_file:
        for i in range(num_configs):
            config_group = hdf5_file[f'config_batch_0/config_{i}']
            properties["atomic_numbers"].append(config_group['atomic_numbers'][:])
            properties["cell"].append(config_group['cell'][:])
            properties["charges"].append(config_group['charges'][:])
            properties["energy"].append(config_group['energy'][()])
            properties["forces"].append(config_group['forces'][:])
            properties["positions"].append(config_group['positions'][:])
            properties["stress"].append(config_group['stress'][:])
            properties["virials"].append(config_group['virials'][:])

    return properties
```

```python
def plot_distribution(data, title):
    plt.figure()
    plt.hist(data, bins=50, alpha=0.75)
    plt.title(title)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
```

```python
def plot_property_distributions(properties):
    plot_distribution([item for sublist in properties["atomic_numbers"] for item in sublist], "Atomic Numbers Distribution")
    plot_distribution([cell_val for cell in properties["cell"] for cell_val in cell.flatten()], "Cell Values Distribution")
    plot_distribution([item for sublist in properties["charges"] for item in sublist], "Charges Distribution")
    plot_distribution(properties["energy"], "Energy Distribution")
    plot_distribution([force_val for forces in properties["forces"] for force in forces for force_val in force], "Forces Distribution")
    plot_distribution([pos_val for positions in properties["positions"] for pos in positions for pos_val in pos], "Positions Distribution")
    plot_distribution([stress_val for stress in properties["stress"] for stress_val in stress.flatten()], "Stress Distribution")
    plot_distribution([virial_val for virials in properties["virials"] for virial in virials for virial_val in virial], "Virials Distribution")
```

```python
file_path = '../datasets/train_0.h5'
properties = read_data(file_path)
plot_property_distributions(properties)
```
